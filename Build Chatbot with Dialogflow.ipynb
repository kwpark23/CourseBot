# Setting up and Training Chatbot on Dialogflow and deploying on Messenger

## 1. The Brief
Imagine if you can build a chatbot using tools that exists already and be able to deploy it on Facebook Messenger for public to use. Won't that be cool?
I'm here to show you how to set up and train a chatbot that is suited for your purpose and connect it to Messenger. Let's get started!

## 2. Setting up Dialogflow account
Dialogflow is a platform for creating multi-purpose chatbots by Google. You don't need coding knowledge to make a chatbot on Dialogflow. 
Go to the link [here](https://dialogflow.com/) and click on `Sign up for free` to create your account. Once you have created your account, sign in with your account. 

## 3. Creating your first chatbot
Click `Go to console` or `Create a new agent` buttons to make your first chatbot.

![Imgur](https://i.imgur.com/8E4DFOc.png)

Dialogflow console is divided up into 3 main sections. Section 1 has all the options for navigating to different pages. Section 2 is where you will interact the console to train your chatbot. Section 3 allows you to test your chatbot after you have trained it. 

Once you are ready to create your chatbot, type in the name of your chatbot in `Agent Name` box and click `Create` button. I have created a new book suggestion chatbot called 'BookBot'.

## 4. Enabling Small Talk
Before you do anything, you want your chatbot to be able to carry out small talk with user. 

![Imgur](https://i.imgur.com/2PmwAZW.png)

First, nagivate to `Small Talk` page from nagivation pane. Second, click on `Enable` button to enable small talk for your chatbot. Finally, click `Save` button to save the change.

## 5. Testing your chatbot
On the far right, you will be able to test your chatbot by typing in phrases. Click on `Try it now` and type in greetings. 

![Imgur](https://i.imgur.com/bRj96cs.png)

Then, you will be able to see what the user typed and your chatbot's default response. 

## 6. Creating Entities
Now, it's time to customize your chatbot for your specific purpose. You want to navigate to `Entities` page and click on `Create Entity`. Entities are used to identifying data correctly from information user provided. For example, if user typed 'I want mushroom on my pizza', mushroom is the topping on the pizza. So, we want to add 'mushroom' to entity called, 'Toppings' to recognize it as pizza topping. We can also add 'pineapple', 'pepperoni', 'pepper' and any other pizza toppings under entities. 

For my BookBot, I am going to create two entities: 'Authors' and 'BookTitles'. I got the list of authors and book titles from Indigo [website](https://www.chapters.indigo.ca/en-ca/books/summer-reading/?ref=shop%3ahome%3ahome-main%3a01-hp-hero-summer-reading%3ahero&cl=01-hp-hero-summer-reading_hero). When you are entering reference values, you want to enter things user might mention when using your chatbot. When your chatbot recognizes those things from user's phrases, it will tag those words with entities you have created. Make sure to enter variation of what user might ask for as synonyms. 

Once you have entered your reference values, click `Save` to train your chatbot. 

![Imgur](https://i.imgur.com/AOS9jb8.png)

![Imgur](https://i.imgur.com/HYlyOHT.png)

## 7. Creating Intents
This is where things get interesting. You want your chatbot to be able to process user's question and reply with appropriate response. Intent is a broad description of what a person is trying to say. For example, if the person asks, 'I'm looking for a Mexican restaurant in the center of town', intent is "restaurant search". 

Go to `Intents` page and you will see a lot of options to choose from. For now, we are only interested in `Training Phrases` and `Responses` sections of the page. Also, if you want specific entity to be present in user's intent, under `Action and parameters`, check `Required` box next to the entity of your choice. 

One of the intents I am going to create is called `Karen Swan`. When the user searches for a latest book by Karen Swan, I want to make a suggestion of her book. In `Training Phrases` section, you want to type in as many phrases as you can think of that a user might ask. In `Responses` sections you can type a list of responses you want to give to user's question.

In `Trainig Phrases` section, you can manually assign an entity by highlighting a part of the phrase and selecting an entity of your choice.

## 8. Testing your chatbot
Finally, the fun part! Now is the time to test if your chatbot is working correctly. Enter a few phrases you have trained your chatbot for and see if you get the responses you expect. 

![Imgur](https://i.imgur.com/x1E3kf0.png)
Here, I typed, 'make me a book suggestion' and got the response, 'How about Broken Throne: A Red Queen Collection by Victoria Aveyard', which is exactly what I trained it for. Seems to be working!

Of course, you can train your chatbot further by adding more entities and intents. More you train it, more diversity in responses your chatbot can have.  

## 9. Integrating with Facebook Messenger
Now, we know our chatbot is working so we want to deploy it on Facebook Messenger. Click on `Integrations` tab from navigation pane and click on Facebook Messenger to turn it on. 

![Imgur](https://i.imgur.com/KZdWKki.png)

When you click on Facebook Messenger link, you will see an instruction on how to deploy it. What we need is a Facebook Page Access Token. 

![Imgur](https://i.imgur.com/W4GR1QE.png)

Time to head over to Facebook Developer website.

## 10. Getting Facebook Page Access Token
Let's head over to Facebook Developer website by clicking this [link](https://developers.facebook.com/). If you already have a Facebook account, just simply log in.  

Now, click on `My Apps` on the top right corner and click `Create an App`.

![Imgur](https://i.imgur.com/GPlijTg.png)

Create the display name of your App and click `Create App ID`. You have to do a quick security check to prove that you are not a robot.

![Imgur](https://i.imgur.com/1Sc0QNx.png)

You will be lead to your bot's dashboard. 

![Imgur](https://i.imgur.com/X5ixdCv.png)

First, click `Skip` for the first page you see like in the image above.

Then, if you scroll down, you will see `Add a Product` section. Click on `Set Up` button for Messenger. 

In the next page you see, scroll down until you see `Access Tokens` section. You don't have Page Access Token yet for your chatbot because you don't have a Facebook page yet. What you want to do is create a new page. 

![Imgur](https://i.imgur.com/6QJecWp.png)

When you click on `Create a New Page`, you will be led to the following page:

![Imgur](https://i.imgur.com/tSz1Hid.png)

Click on either options to create a page. I have decided to create my BookBot page as a community page but it is up to you. 

![Imgur](https://i.imgur.com/FAXRtUx.png)

Once you click through a few options for profile picture, you will finally be able to see your bot's Facebook page. 

![Imgur](https://i.imgur.com/E9sVpEs.png)

Now, time to get back to Facebook Developer page for your bot. When you refresh the page, you can now select your bot's page to get Access Token. But you are not quite there yet. 

![Imgur](https://i.imgur.com/S9pb3Kp.png)

You will be told that you have to edit your permission to get access token. Just click on `Edit Permissions` and follow along the instruction until your bot is linked to the Facebook page. 

Now, you will see Page Access Token appear. 

![Imgur](https://i.imgur.com/LqoOa5W.png)

Click on it to copy to a clipboard and head over to Dialogflow page. 

![Imgur](https://i.imgur.com/NuWrOcJ.png)

## 11. Setting up your bot in Dialogflow

Paste the Access Token and write anything for `Verify Token`. It can be any word or phrase you choose. Once you are ready, click `Start`.

![Imgur](https://i.imgur.com/Qu8BMF5.png)

You will see two things: `Start` have changed to `Stop` and a green icon will appear, saying 'Agent Training Completed'. Now click on the clipboard icon for `Callback URL` and head back over to Facebook Developer page for your bot. 


## 12. Hooking your bot from Dialogflow to Facebook

![Imgur](https://i.imgur.com/htsWHTb.png)

Right under Access Token section, you will see `Webhook` section. Click on `Subscribe to Events`

![Imgur](https://i.imgur.com/24KwQnA.png)

A couple of things to do here. First, paste `Callback URL` and fill in `Verify Token`. Then, choose a subscription field (you can choose more than one) and finally click `Verify and Save` button.

![Imgur](https://i.imgur.com/hCArdlN.jpg)

Lastly, you want to choose the page you want to subscribe to and click `Subscribe`. 

Now you are all set! How do you know?

![Imgur](https://i.imgur.com/EHxCGbS.jpg)

On the navigation pane to the left, if you see the word `Webhook` with a green circle and a check mark, that means all the set up has been successfully completed. Now head over to your bots Facebook page.

## 13. Testing out your bot on Messenger

Once you are over to your bot's Facebook page, click on three dot icons and click `View as Page Visitor`.

![Imgur](https://i.imgur.com/oFDOtGV.jpg)

Now the moment of the truth! Click on `Send Message` and when the chatbox pops up start tying in question you have trained your bot for!

![Imgur](https://i.imgur.com/LEgUYae.jpg)

Open the chatbox in Messenger mode and here is what you will see.


![Imgur](https://i.imgur.com/d6jt4QO.jpg)

Your bot is answering your questions just like you trained it to be! Isn't this cool?
